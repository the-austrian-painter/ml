```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

IceCream = pd.read_csv('/content/drive/MyDrive/ml-practical/datasets/IceCream.csv')
IceCream.head(10)
```

```python
sns.jointplot(x='Temperature',y='Revenue',data=IceCream)
```

```python
y=IceCream['Revenue']
```

```python
x=IceCream[['Temperature']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size = 0.25)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor = regressor.fit(x_train,y_train)
```

```python
print('linear model coefficient (m): ', regressor.coef_)
print('linear model intercept (b): ', regressor.intercept_)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_test
```

```python
plt.scatter(x_train,y_train,color = 'green')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')
plt.ylabel('Revenue(dollars)')
plt.xlabel('Temperature(degC)')
plt.title('Revenue vs Temperature')
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.values.reshape(len(y_test),1)), 1))
```
